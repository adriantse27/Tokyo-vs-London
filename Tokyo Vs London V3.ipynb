# 1. When would be the perfect time to book airbnb?

```python
# creating copies of the calendars
availability_Tokyo = TokyoCalendar.copy()
availability_London  = LondonCalendar.copy()

# calculating mean listing availability grouped by date
availability_Tokyo = availability_Tokyo.groupby('date')['booked'].mean().reset_index()
availability_Tokyo['date'] = pd.to_datetime(availability_Tokyo['date'])

availability_London = availability_London.groupby('date')['booked'].mean().reset_index()
availability_London['date'] = pd.to_datetime(availability_London['date'])
```

From the above chart, we see that London Airbnb listing is highly popular throughout the year. Tokyo however is heavily declined from Oct 2020 to early 2021, but it is greatly increasing in 2021 probably due to the Pfizer and Moderna vaccines being announced.
```python
def available_listings (df):
    """
    Description: This function is used to select all the listings that are currently listed as available(not booked).

    Arguments:
        Data Frame that you will be querying

    Returns:
        Data Frame Has Only Available Listings and its variables and other fields.
    """
    df1 = df.query('booked == 0')
    return (df1)
```

# Data Modelling

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
#Price is the dependent varaible which is already in y, so it needs to be dropped from X and leave out other columns
X=df2.drop('GBPPrice', axis=1)
y= df2['GBPPrice']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)


regression=LinearRegression()
regression.fit(X_train,y_train)

y_pred=regression.predict(X_test)

score = r2_score(y_test, y_pred)
score
```

The score is very low, this means that the room type has almost no correlation to how much it will cost as we are unable to predict its values based on what they are. There must be other factors that go into how much things costs.

```python
def DataModelling (X,y,tsize):
    """
    Description: This function performs basic linear regression using the dependent and independent variables stored into two dataframes and also the test size.

    Arguments:
        Pass the indep, dep, test size into the parameters

    Returns:
        Returns a test score that indicates how reliable the predicted values are.
    """
    #Split
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=tsize)
    
    #Fit
    regression=LinearRegression()
    regression.fit(X_train,y_train)

    #Predict
    y_pred=regression.predict(X_test)
    
    #Score
    score = r2_score(y_test, y_pred)
    return (score)
```
